```python
api_key=os.getenv('OPENAI_API_KEY')
```

```python
if api_key:
    print('ture')
```

```python
message=[]
```

```python
message.append({'role':'system','content':'You are a financial expert specializing in real estate investment and negotiation'})
```

```python
message
```

```python
client=openai.OpenAI(api_key=api_key)
```

```python
client
```

```python
def chatGPTresponse(user_message):
    message.append({'role':'user','content':user_message})
    response=client.chat.completions.create(model="gpt-3.5-turbo",
        messages=message
    )
    reply=response.choices[0].message.content
    message.append({'role':'assistant','content': reply})
    
    return reply
```

```python
# !pip install gradio
```

```python
import gradio as gr
```

```python
demo = gr.Interface(fn=chatGPTresponse, 
                    inputs="text", 
                    outputs="text", 
                    title="My Saving Assistant")
```

```python
demo.launch(share=True)
```
