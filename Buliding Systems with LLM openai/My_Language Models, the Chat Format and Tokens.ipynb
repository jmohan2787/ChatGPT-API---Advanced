```python
import openai
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
api_key=os.getenv("OPENAI_API_KEY")
```

```python
client=openai.OpenAI(api_key=api_key)
```

```python
def get_completion(prompt,model='gpt-3.5-turbo'):
    client=openai.OpenAI(api_key=api_key)
    response=client.chat.completions.create(model=model, messages=[{'role':'user', 'content':prompt}],temperature=0,)
    return response.choices[0].message.content
```

```python
get_completion('what is captital of pairs?')
```

```python
response = get_completion("Take the letters in lollipop \
and reverse them")
print(response)
```

```python
def get_completion_from_message(messages,model='gpt-3.5-turbo',temperature=0,max_tokens=500):
    client=openai.OpenAI(api_key=api_key)
    response=client.chat.completions.create(model=model,messages=messages,temperature=temperature,max_tokens=max_tokens)
    return response.choices[0].message.content
```

```python
messages = [
    {'role': 'system', 'content': "You are an assistant who responds in the style of Dr Seuss."},
    {'role': 'user', 'content': "Write me a very short poem about a happy carrot"}
]
get_completion_from_message(messages,temperature=1)
```

```python
# length
messages =  [
{'role':'system',
 'content':'All your responses must be \
one sentence long.'},
{'role':'user',
 'content':'write me a story about a happy carrot'},
]
response = get_completion_from_message(messages, temperature =1)
print(response)
```

```python
# combined
messages =  [
{'role':'system',
 'content':"""You are an assistant who \
responds in the style of Dr Seuss. \
All your responses must be one sentence long."""},
{'role':'user',
 'content':"""write me a story about a happy carrot"""},
]
response = get_completion_from_message(messages,
                                        temperature =1)
print(response)
```

```python
def get_completion_and_token_count(messages,
                                   model="gpt-3.5-turbo",
                                   temperature=0,
                                   max_tokens=500):
    client=openai.OpenAI(api_key=api_key)
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens,
    )

    content = response.choices[0].message.content  # Updated way to access content

    token_dict = {
        'prompt_tokens': response.usage.prompt_tokens,  # Updated way to access token counts
        'completion_tokens': response.usage.completion_tokens,
        'total_tokens': response.usage.total_tokens,
    }

    return content, token_dict
```

```python
messages = [
    {'role': 'system', 'content': "You are an assistant who responds in the style of Dr Seuss."},
    {'role': 'user', 'content': "Write me a very short poem about a happy carrot"}
]

response, token_dict = get_completion_and_token_count(messages)

print("Response:", response)
print("Token Usage:", token_dict)
```
